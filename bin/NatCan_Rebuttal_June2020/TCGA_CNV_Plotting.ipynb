```r
library(ggplot2)
library(ggpubr)

setwd("~/Desktop/H4H/pughlab/projects/BTSCs_scRNAseq/Manuscript_G607removed/NatCan_Rebuttal/CNV_TCGA/")
```

```r
### load data
med <- readRDS("TCGA_GBM_CNV_armMed.rds")
med <- t(med)

avg <- readRDS("TCGA_GBM_CNV_armAvg.rds")
avg <- t(avg)
avg <- round(avg) #round to neareast number

mode <- readRDS("TCGA_GBM_CNV_armMode.rds")
mode <- t(mode)

counts <- readRDS("TCGA_GBM_CNV_Counts.rds")

gsva <- readRDS("TCGA_GBM_RNA_GSVA.rds")

### Med and Avg have 5718 of the same calls
### Med and Mode have 5760 of the same calls
### Avg and Mode have 5706 of the same calls
### So pretty much the same results no matter how you slice it
```

```r
### merge rounded avg results with GSVA scores
plot.dat <- cbind(gsva, avg)
#head(plot.dat)
plot.dat[plot.dat == 0] <- "Neutral"
plot.dat[plot.dat ==  -1] <- "Deletion"
plot.dat[plot.dat ==  1] <- "Gain"
#head(plot.dat)

dev <- plot.dat[, c(1, 3:ncol(plot.dat))]
colnames(dev)[1] <- "Score"
dev$GSC_Score <- "Developmental"

IR <- plot.dat[, c(2:ncol(plot.dat))]
colnames(IR)[1] <- "Score"
IR$GSC_Score <- "InjuryResponse"
plot.dat <- rbind(dev, IR)
GSC_Score <- plot.dat$GSC_Score

plot.dat <- as.data.frame(lapply(plot.dat, function (x) if (is.character(x)) factor(x, levels = c("Deletion", "Neutral", "Gain"), ordered = T) else x)) 
plot.dat$GSC_Score <- GSC_Score
#str(plot.dat)
head(plot.dat)

#str(plot.dat)
```

```r
### need to refactor analysis
### paired samples for t-test in this case

p <- ggboxplot(plot.dat, x = "chr1p", y = "Score",
          color = "GSC_Score", palette = c("red", "black"),
          add = "jitter") + ggtitle("chr19p") + xlab("") + 
          ylab("GSVA Score")
p + stat_compare_means(aes(group = GSC_Score), 
                       label = "p.format", 
                       paired = T
                      )
```

### Bin tumours into Dev and IR
----

Bin values into yes and no.
Take top 25% of data to be Dev vs IR (upper quartile)

t test was used to compare the bulk scores between all tumors with and those without that. 

```r
plot.dat <- cbind(gsva, avg)
#head(plot.dat)
plot.dat[plot.dat == 0] <- "Neutral"
plot.dat[plot.dat ==  -1] <- "Deletion"
plot.dat[plot.dat ==  1] <- "Gain"
head(plot.dat)
```

```r
quantile(plot.dat$Developmental_GSC) #take top 25% of points
quantile(plot.dat$InjuryResponse_GSC)

pdf("TCGA_GBM_GSVA_Scores_Quartiles.pdf", width = 6, height = 6)
plot(plot.dat$Developmental_GSC,
     plot.dat$InjuryResponse_GSC,
     xlab = "Developmental GSVA Score",
     ylab = "Injury Response GSVA Score",
     main = "TCGA GBM (n=148)"
    )
abline(h=0.185, v = 0.186, lty = 2)
dev.off()
```

```r
Dev.tumours <- plot.dat[round(plot.dat$Developmental_GSC, 3) >= 0.185 ,]
nonDev.tumours <- plot.dat[!round(plot.dat$Developmental_GSC, 3) >= 0.185 ,]

IR.tumours <- plot.dat[round(plot.dat$InjuryResponse_GSC, 3) >= 0.186 ,]
nonIR.tumours <- plot.dat[!round(plot.dat$InjuryResponse_GSC, 3) >= 0.186 ,]
```

```r
### chr6q deletion enrichment in Dev tumours

table(Dev.tumours$chr6q)
table(nonDev.tumours$chr6q)

Dev_del_6q <- matrix(c(10,27,39,72), nrow = 2, ncol = 2)
Dev_del_6q
fisher.test(Dev_del_6q)
```

```r
### chr9p gain enrichment in Dev tumours

table(Dev.tumours$chr9p)
table(nonDev.tumours$chr9p)

Dev_gain_9p <- matrix(c(0,37,11,100), nrow = 2, ncol = 2)
Dev_gain_9p
fisher.test(Dev_gain_9p)
```

```r
### chr9p gain enrichment in Dev tumours

table(Dev.tumours$chr19p)
table(nonDev.tumours$chr19p)

Dev_gain_19p <- matrix(c(13,24,36,75), nrow = 2, ncol = 2)
Dev_gain_19p
fisher.test(Dev_gain_19p)
```
