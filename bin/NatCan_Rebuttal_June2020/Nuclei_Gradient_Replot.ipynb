```r
library(ggplot2)
library(ggpubr)
```

```r
combo.dat <- readRDS("~/Desktop/H4H/pughlab/projects/BTSCs_scRNAseq/Manuscript_G607removed/NatCan_Rebuttal/Nuclei_rescore/GSC_GBM_NuclieLive_Meta.rds")
```

```r
combo.dat$SampleTech <- paste(combo.dat$SampleType, combo.dat$Technology, sep = "_")
```

```r

cells <- combo.dat[combo.dat$Technology == "LiveCell", ]
nuc <- combo.dat[combo.dat$Technology == "Nuclei", ]
```

```r
dim(cells)
```

```r
a <- ggscatterhist(cells, 
              x = "Developmental_AUC", 
              y = "InjuryResponse_AUC",
              palette = c("#225ea8", "black", "red"), 
              color = "SampleTech",
              size = 0.5,
              alpha = 0.1,
              #ellipse = TRUE, 
              #ellipse.type = "confidence",
              margin.params = list(fill = "SampleTech", color = "black", size = 0.2),
              ggtheme = theme_classic(),
              #ellipse.alpha = 0
              #ylim = c(0.1,0.45),
              #xlim = c(0.05,0.4)
             ) 
```

```r
b <- ggscatterhist(nuc, 
              x = "Developmental_AUC", 
              y = "InjuryResponse_AUC",
              palette = c("darkred"), 
              color = "darkred",
              size = 0.5,
              alpha = 0.1,
              #ellipse = TRUE, 
              #ellipse.type = "confidence",
              margin.params = list(fill = "SampleTech", color = "black", size = 0.2),
              ggtheme = theme_classic(),
              ellipse.alpha = 0,
              #ylim = c(0.1,0.45),
              #xlim = c(0.1,0.4)
             ) 
```

```r
pdf("~/Desktop/H4H/pughlab/projects/BTSCs_scRNAseq/Manuscript_G607removed/NatCan_Rebuttal/Nuclei_rescore/Nuc_sep_gradient.pdf", width = 5, height = 5)
a
b
dev.off()
```
