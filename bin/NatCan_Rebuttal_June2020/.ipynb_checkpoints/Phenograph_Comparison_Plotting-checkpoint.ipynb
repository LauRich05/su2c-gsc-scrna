## 1.0 Number of Clusters
----

```r
meta <- readRDS("GSC_Rphenograph_meta.rds")
#meta <- do.call(rbind, meta)
colnames(meta[[1]])
```

```r
### count clusters between methods
pheno_clust_count <- unlist(lapply(meta, function(x) length(unique(x$Rphenograph_clusters))))
seurat_clust_count <- unlist(lapply(meta, function(x) length(unique(x$Cluster.ID))))
clust.dat <- data.frame(cbind(pheno_clust_count, seurat_clust_count))
clust.dat$name <- rownames(clust.dat)
head(clust.dat)
```

```r
pdf("ClusterCounts.pdf", height = 7, width = 7)
ggscatter(clust.dat, 
          x = "seurat_clust_count", 
          y = "pheno_clust_count",
          #color = "cyl", 
          #palette = c("#00AFBB", "#E7B800", "#FC4E07"),
          label = "name", 
          repel = TRUE,
          xlab = "Seurat Clusters # (Original)",
          ylab = "Phenograph Cluster #",
          #xlim = c(1,7),
          #ylim = c(1,13)
         )
dev.off()
```

### Now plot as histogram to stack on other plots below

```r
pheno_clust_count <- unlist(lapply(meta, function(x) length(unique(x$Rphenograph_clusters))))
pheno_clust_count <- as.data.frame(pheno_clust_count)
colnames(pheno_clust_count) <- "cluster"
pheno_clust_count$sample <- rownames(pheno_clust_count)
pheno_clust_count$method <- "phenograph"
#head(pheno_clust_count)

seurat_clust_count <- unlist(lapply(meta, function(x) length(unique(x$Cluster.ID))))
seurat_clust_count <- as.data.frame(seurat_clust_count )
colnames(seurat_clust_count ) <- "cluster"
seurat_clust_count$sample <- rownames(seurat_clust_count )
seurat_clust_count$method <- "seurat"
#head(seurat_clust_count )

clust_hist <- rbind(seurat_clust_count, pheno_clust_count)
```

```r
cluster.plot <- ggplot(clust_hist, aes(sample, cluster)) +
      geom_bar(aes(fill = method), position = "dodge", stat="identity", width=.5) +
      geom_text(aes(label=cluster, group=method), position=position_dodge(width=0.5), vjust=-0.5) + ylim(c(0,15)) + 
      theme_bw(base_size = 18) + theme_classic() +
      ylab('Cluster Number') +  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) + xlab("") 
cluster.plot
```

## 2.0 DE genes / cluster
----

```r
### load DE data
phenoDE <- readRDS("GSC_Rphenograph_DE_PhenographClusters.rds")
seuratDE <- readRDS("GSC_Rphenograph_DE_SeuratClusters.rds")
```

```r
pheno_de.count <- list()

for (i in 1:length(names(phenoDE))){
    

    sample <- names(phenoDE)[i]
    phenoDE[[i]]$sample <- sample
    #print(dim(phenoDE[[i]]))
    phenoDE[[i]] <- phenoDE[[i]][phenoDE[[i]]$p_val_adj < 0.05, ]
    #print(dim(phenoDE[[i]]))
    pheno_de.count[[sample]] <- as.numeric(table(phenoDE[[i]]$cluster))
    pheno_de.count[[sample]] <- as.data.frame(pheno_de.count[[sample]])
    colnames(pheno_de.count[[sample]]) <- "MarkerGenes"
    pheno_de.count[[sample]]$sample <- sample
    pheno_de.count[[sample]]$method <- "phenograph"
}


pheno_de.count <- do.call(rbind, pheno_de.count)

seurat_de.count <- list()

for (i in 1:length(names(seuratDE))){
    

    sample <- names(seuratDE)[i]
    seuratDE[[i]]$sample <- sample
    #print(dim(phenoDE[[i]]))
    seuratDE[[i]] <- seuratDE[[i]][seuratDE[[i]]$p_val_adj < 0.05, ]
    #print(dim(phenoDE[[i]]))
    seurat_de.count[[sample]] <- as.numeric(table(seuratDE[[i]]$cluster))
    seurat_de.count[[sample]] <- as.data.frame(seurat_de.count[[sample]])
    colnames(seurat_de.count[[sample]]) <- "MarkerGenes"
    seurat_de.count[[sample]]$sample <- sample
    seurat_de.count[[sample]]$method <- "seurat"
}

seurat_de.count <- do.call(rbind, seurat_de.count)
```

```r
plot.dat <- rbind(pheno_de.count, seurat_de.count)
tail(plot.dat)
```

```r
#pdf("MarkerDE.pdf", width = 11, height = 5)
DE.plot <- ggplot(plot.dat, 
       aes(x=sample, y=MarkerGenes, fill=method)) + geom_boxplot() +
        theme_classic() + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +
    ylab("Marker Genes / Cluster (FDR < 0.05)") + xlab("")
DE.plot
#dev.off()
```

## 3.0 Sil width / cluster
---

```r
### load DE data
phenoSil <- readRDS("GSC_Rphenograph_Sil_PhenographClusters.rds")
seuratSil <- readRDS("GSC_Rphenograph_Sil_SeuratClusters.rds")
```

```r
seurat_silClust <- list()

for (i in 1:length(names(seuratSil))){  
    sample <- names(seuratSil)[i]
    seurat_silClust[[sample]] <- as.data.frame(seuratSil[[sample]])
    colnames(seurat_silClust[[sample]]) <- "SilWidth"
    seurat_silClust[[sample]]$sample <- sample
    seurat_silClust[[sample]]$method <- "seurat"
}

seurat_silClust <- do.call(rbind, seurat_silClust)

pheno_silClust <- list()

for (i in 1:length(names(phenoSil))){  
    sample <- names(phenoSil)[i]
    pheno_silClust[[sample]] <- as.data.frame(phenoSil[[sample]])
    colnames(pheno_silClust[[sample]]) <- "SilWidth"
    pheno_silClust[[sample]]$sample <- sample
    pheno_silClust[[sample]]$method <- "phenograph"
}

pheno_silClust <- do.call(rbind, pheno_silClust)

plot.sil <- rbind(pheno_silClust, seurat_silClust)
head(plot.sil)
```

```r
sil.plot <- ggplot(plot.sil, 
       aes(x=sample, y=SilWidth, fill=method)) + geom_boxplot() +
        theme_classic() + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +
    ylab("Cluster Silhouette Width") + xlab("")

sil.plot
```

### Arrange plots into one

```r
pdf("MethodComparison_Clust_DE_Sil.pdf", width = 9, height = 11)
par(mfrow=c(3,1))
print(cluster.plot)
print(DE.plot)
print(sil.plot)
dev.off()
```

```r
pdf("MethodComparison_Clust_DE_Sil.pdf", width = 9, height = 11)
ggarrange(cluster.plot, DE.plot, sil.plot, 
          common.legend = T,
          ncol = 1,
          nrow = 3
         )
dev.off()
```
