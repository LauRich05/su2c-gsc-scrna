## 3.0 Compare gene loadings
---

---
### 3.1 Correlate loadings
---

```r
scaled.pca <- readRDS("GBM_GSC_SplitScaled_PCA.rds")
orig.pca <- readRDS("GBM_GSC_Original_PCA.rds")
```

```r
### isolate and merge gene loadings
scaled.genes <- scaled.pca@feature.loadings
orig.genes <- orig.pca@feature.loadings

genes <- cbind(orig.genes[rownames(scaled.genes), ], 
               scaled.genes[ ,1:50]
              )
#head(genes)

### correlate gene loadings 
### to identify similar PCs between methods
corr.mat <- cor(genes, method = "pearson")
corr.mat <- data.frame(corr.mat)
```

```r
### pretty low correlation across everything
pheatmap(corr.mat,
         file = "Scaled_Orig_GeneLoadings_Correlation.pdf",
         width = 10, 
         height = 10,
         border_color = NA,
         fontsize = 4
        )
```

```r
### what are the top correlated scaled equivs in orig

corr.mat[order(corr.mat[ ,"PC_2"]), ]
```

```r
scaled.genes[order(scaled.genes[ ,"ScaledPC_5"]), ]
```

### 3.2 Correlate gene loadings and gene sig scores
---

```r
sigs <- meta[ ,grepl("AUC", colnames(meta))]
sigs <- cbind(sigs, scaled.pca@cell.embeddings)
#head(sigs)

### correlate gene signatures and PC loadings
sigs.cor <- cor(sigs, method = "spearman")
sigs.cor <- sigs.cor[grepl("ScaledPC", colnames(sigs.cor)), grepl("AUC", rownames(sigs.cor))]
head(sigs.cor)
```
